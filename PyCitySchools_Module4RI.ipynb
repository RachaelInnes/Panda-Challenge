    Analysis was undertaken using data from Local Area government schools on a number of variables, such as School Type (Government or Independant),  Budget, Academic peformance (Maths and Reading Pass rates) and size (Number of students) to conduct draw insights into the relationships that may exist between these factors. At a high level, the analysis shows:
  
  -The overall pass rate for all schools in the Local Government Area is 72.8%.
  -Bailey High Shcool has the highest overall pass rate of 80.04%, this school also has the largest student population (4976).
  -In terms of academic performance, there are four Independant schools in the Top 5, and one government school of the %of Overall Passing students (Maths and Reading scores at 50% or above) and four government schools and one Indpedant schools in the bottom 5 of this group.
  -In terms of Reading and Maths scores by Year Level across Schools, students tend to peform consistently in both Reading and Maths across each year.
  -There is little difference between the size of the school and the academic perforamnce, all are fairly similar (7 - 74% overall passing rate).
  
There are two conclusions that could be drawn from this data, as follows:

1. The Type of school does correlate with academic performance. From this data there is higher overall passing rates in Indpendant Schools (76.97%) than Government Schools (70.69%). This seems to be attributed to higher levels of passing in the ratio of of students in Independant Schools passing maths).

2. The overall Budget and Per Student Budget does not have a correlation with academic performance (% Overall Passing Rates), this is obvious where the Top 2 Performing Schools have very different budgets. That is Bailey High School with a budget of 3.1m compared to Carbera with just over 1m budget.

```python
#Total student by School
per_school_counts = school_data_complete_df["school_name"].value_counts()
print("Total Student Count per School:")
print(per_school_counts)

#Calculate the total budget per school
per_school_budget = school_data_complete_df.groupby("school_name")["budget"].mean()
print("\nTotal School Budget per School:")
print(per_school_budget)

# Calculate the school budget per student (per school budget)
per_student_budget = per_school_budget / per_school_counts
print("\nPer Student Budget:")
print(per_student_budget)

#Calculate average maths score 
per_school_maths = school_data_complete_df.groupby("school_name")["maths_score"].mean()
print("\nAverage Math Scores per School:")
print(per_school_maths)

#Calculate the average reading test score
per_school_reading = school_data_complete_df.groupby("school_name")["reading_score"].mean()
print("\nAverage Reading Scores per School:")
print(per_school_reading)
```

```python
# Get the students who passed maths and passed reading by creating separate filtered DataFrames from school_data_complete.
passed_math = school_data_complete_df[school_data_complete_df['maths_score'] >= 50]

passed_reading = school_data_complete_df[school_data_complete_df['reading_score'] >= 50]

# Get the students who passed both reading and maths in a separate DataFrame from school_data_complete.
passing_maths_and_reading = school_data_complete_df[(school_data_complete_df['maths_score'] >= 50) & (school_data_complete_df['reading_score'] >= 50)]
```

```python
#Calculate the percentage of students passing math for each school
per_school_passing_maths = (passed_math.groupby('school_name').size() / per_school_counts) * 100
per_school_passing_maths

# Calculate the percentage of students passing reading for each school
per_school_passing_reading = (passed_reading.groupby('school_name').size() / per_school_counts) * 100
per_school_passing_reading

# Calculate the percentage of students passing both math and reading for each school
per_school_passing_both = (passing_maths_and_reading.groupby('school_name').size() / per_school_counts) * 100
per_school_passing_both

print("Percentage of students passing math for each school:")
print(per_school_passing_maths)

print("Percentage of students passing reading for each school:")
print(per_school_passing_reading)

print("Percentage of students passing both math and reading for each school:")
print(per_school_passing_both)
```

## Maths Scores by Year

```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# Read the data from a CSV file into a DataFrame
school_data_complete = pd.read_csv("Resources/students_complete.csv")

# Create data series of scores by year levels using conditionals
year_nine = school_data_complete[(school_data_complete["year"] == 9)]
year_ten = school_data_complete[(school_data_complete["year"] == 10)]
year_eleven = school_data_complete[(school_data_complete["year"] == 11)]
year_twelve = school_data_complete[(school_data_complete["year"] == 12)]

# Group each by school name
year_nine_scores = year_nine.groupby("school_name")["maths_score"].mean()
year_ten_scores = year_ten.groupby("school_name")["maths_score"].mean()
year_eleven_scores = year_eleven.groupby("school_name")["maths_score"].mean()
year_twelve_scores = year_twelve.groupby("school_name")["maths_score"].mean()

# Combine series into single DataFrame
maths_scores_by_year = pd.DataFrame({
    "9th Grade": year_nine_scores,
    "10th Grade": year_ten_scores,
    "11th Grade": year_eleven_scores,
    "12th Grade": year_twelve_scores
})

# Minor data wrangling
maths_scores_by_year.index.name = None

styled_df = maths_scores_by_year.style.set_properties(**{
                'text-align': 'center',
                'border-color': 'black',
                'border-style': 'solid',
                'border-width': '1px'
            }).set_table_styles([{
                'selector': 'th',
                'props': [('font-weight', 'bold')]
            }])

# Display the styled DataFrame
styled_df
```
